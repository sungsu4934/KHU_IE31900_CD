```python
from lxml.html import fromstring

import time
import pandas as pd
import requests
```

```python
# 일단 나이와 이름이 동일한 사람은 "이상훈(1971)"만 파악되었음
hitter = pd.read_csv('선수데이터(전처리완료)/타자.csv')
pitcher = pd.read_csv('선수데이터(전처리완료)/투수.csv')
```

# 기사 데이터수집
 - 키워드: 야구 + 선수명 + (선수나이)
   - ex) 야구 + 이종범 + (29)

```python
# 타자

t1 = time.time()
player_num = 0
hitter_info = pd.DataFrame()

for idx in range(hitter.shape[0]):

    # 값 초기화
    page = 1
    news_count = 0

    ID = hitter.loc[idx, 'ID']
    name = hitter.loc[idx, '선수명']
    age = int(hitter.loc[idx, '연도'] - int(hitter.loc[idx, '생년월일'].split('-')[0]))
    year = int(hitter.loc[idx, '연도'])

    # 크롤링
    while True:

        ### url 호출
        url = f'https://search.naver.com/search.naver?where=news&query=야구+%20%2B%20{name}%20%2B%20({age})&sm=tab_opt&sort=0&photo=0&field=0&pd=3&ds={year}.01.01&de={year}.12.31&docid=&related=0&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so%3Ar%2Cp%3Afrom{year}0101to{year}1231&is_sug_officeid=0&start={(page-1)*10 + 1}'
        res = requests.get(url)
        parser = fromstring(res.text)
        number_of_news = len(parser.xpath('/html/body/div[3]/div[2]/div/div[1]/section/div/div[2]/ul/li'))

        ### 조기종료 조건: 페이지를 넘겼을 때, 뉴스가 10개가 아닌경우
        if number_of_news != 10:
            news_count += number_of_news
            player_num += 1
            
            ##### 데이터프레임 추가
            tmp = pd.DataFrame({'ID':[ID], '선수명':[name], '연도':[year], '뉴스개수':[news_count]})
            hitter_info = pd.concat([hitter_info, tmp]).reset_index(drop=True)
            
            ##### 잠시 휴식
            if player_num % 100 == 0:
                print(f'{player_num}/{hitter.shape[0]}번째 선수까지 완료!! == 경과시간: {round(time.time()-t1)}초 == 10초 휴식')
                time.sleep(10)

            break
        else:
            news_count += 10
            page += 1        
```
