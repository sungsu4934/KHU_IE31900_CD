# Load Package

```python
from lxml.html import fromstring

import pandas as pd
import numpy as np

import requests
import time
import os
```

# 데이터 불러오기

```python
hitter = pd.read_csv('선수데이터(전처리완료)/타자.csv')[['선수명', '생년월일']].drop_duplicates()
pitcher = pd.read_csv('선수데이터(전처리완료)/투수.csv')[['선수명', '생년월일']].drop_duplicates()
df = pd.concat([hitter, pitcher]).reset_index(drop=True)
df
```

```python
PLAYER_NAMES = list(df['선수명'])
BIRTHDAYS = list(df['생년월일'])
```

# Crawling

```python
cnt = 0
t1 = time.time()

for PLAYER_NAME, BIRTHDAY in zip(PLAYER_NAMES, BIRTHDAYS):

    # 선수 count용
    cnt += 1
    tmp = pd.DataFrame()
    
    # 1차 url을 대입하여 수집
    url = f'http://www.statiz.co.kr/player.php?opt=11&name={PLAYER_NAME}&birth={BIRTHDAY}'
    res = requests.get(url)
    parser = fromstring(res.text)

    element_nums = len(parser.xpath(f'/html/body/div[1]/div[1]/div/section[2]/div/div[2]/div/div/div/div[2]/table/tr/td[1]'))
    years = [parser.xpath(f'/html/body/div[1]/div[1]/div/section[2]/div/div[2]/div/div/div/div[2]/table/tr[{element_num}]/td[1]')[0].text_content() for element_num in range(2, element_nums+2)]
    salarys = [parser.xpath(f'/html/body/div[1]/div[1]/div/section[2]/div/div[2]/div/div/div/div[2]/table/tr[{element_num}]/td[2]')[0].text_content() for element_num in range(2, element_nums+2)]
    wars = [parser.xpath(f'/html/body/div[1]/div[1]/div/section[2]/div/div[2]/div/div/div/div[2]/table/tr[{element_num}]/td[3]')[0].text_content() for element_num in range(2, element_nums+2)]

    tmp = pd.DataFrame({'선수명':[PLAYER_NAME] * element_nums,
                        '생년월일':[BIRTHDAY] * element_nums,
                       '연도':years,
                       '연봉':salarys,
                       'WAR':wars})

    if cnt == 1:
        final = tmp.copy()
    else:
        final = pd.concat([final, tmp]).reset_index(drop=True)
    
    
    if cnt % 100 == 0:
        print(f'{cnt}명 완료 == 경과시간: {round(time.time()-t1)}초 == 10초 휴식')
        time.sleep(10)
```
