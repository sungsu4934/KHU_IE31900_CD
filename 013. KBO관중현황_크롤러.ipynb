```python
from lxml.html import fromstring

import pandas as pd
import requests
```

# 관중현황 데이터수집

```python
url = 'https://www.koreabaseball.com/History/Crowd/History.aspx'
res = requests.get(url)
parser = fromstring(res.text)
colnames = [parser.xpath(f'//*[@id="tblHistory"]/thead/tr/th[{num}]')[0].text_content() for num in range(1, 14)]

all_row = []

for idx in range(1, 20):
    
    # element값 초기화
    row = []
    
    # element값 수집
    for num in range(1, 14):

        if num == 1:
            element = parser.xpath(f'//*[@id="tblHistory"]/tbody/tr[{idx}]/td[{num}]')[0].text_content()
            row.append(element)
            continue

        try:
            element = parser.xpath(f'//*[@id="tblHistory"]/tbody/tr[{idx}]/td[{num}]')[0].text_content().split('(')[1].split(')')[0]
        except:
            element = ''

        row.append(element)
    
    # 연도별로 추가
    all_row.append(row)
    
print('수집완료!!')
```

```python
df = pd.DataFrame(all_row, columns = colnames)
df
```

```python
df.to_csv('관중현황데이터/관중현황.csv', index=False, encoding='utf-8-sig')
```
