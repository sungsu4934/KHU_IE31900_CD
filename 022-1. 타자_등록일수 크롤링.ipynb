# 크롤링

```python
INDEXLIST_OF_YEARS = [num for num in range(40,39,-1)] # (39,20,-1)
INDEXLIST_OF_YEARS
```

```python
t1 = time.time()

########################## 접속과정 ##########################
# page에 접속
driver = webdriver.Chrome('chromedriver_96.exe', chrome_options=options) # 크롬 드라이버 다운로드후 본인 경로로 줍니다.
base_url = 'https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx'
driver.get(base_url)
time.sleep(2)
driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[2]')[0].click()
time.sleep(2)
driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[2]/select/option[1]')[0].click()
time.sleep(2)



########################## 연도 변경과정 ##########################
# 년도 바꾸기 (option[21] ~ option[39]) 2002~2020
for INDEX_YEAR in INDEXLIST_OF_YEARS: 
    
    ### 기본기록 클릭 (세부정보에서 년도 바꾸기가 먼저 눌리면 팀 개수 구할 때 반영이 안되는 이슈가 존재)
    try:
        driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/ul/li[1]/a')[0].click()
    except:
        pass
    
    ### 년도 콤보박스 클릭
    driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[1]/select')[0].click()
    time.sleep(1)
    
    ### 세부년도 클릭
    BOX_YEAR = driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[1]/select/option[{INDEX_YEAR}]')[0]
    BOX_YEAR_PATH = BOX_YEAR.text
    BOX_YEAR.click() # 2020년이 options[39]
    time.sleep(1)
    
    ### 그 해 팀의 개수 파악
    INDEXLIST_OF_TEAMS = len(driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[3]/select/option')) - 1


    ########################## 팀 변경과정 ##########################
    # 팀바꾸기 (option[2] ~ 그 시즌 팀의 개수)
    for INDEX_TEAM in range(2,INDEXLIST_OF_TEAMS+2):
        
        ### 팀 콤보박스 클릭
        driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[3]/select')[0].click()
        time.sleep(1)

        ### 세부 팀 클릭
        BOX_TEAM = driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[3]/select/option[{INDEX_TEAM}]')[0]
        BOX_TEAM_PATH = BOX_TEAM.text
        BOX_TEAM.click() # option[2]부터 팀이름 나옴(1은 팀선택)
        time.sleep(1)
        
        ### 기본기록 클릭
        try: # 첫 수집 시 페이지가 1로 안 바뀌는 것을 보정하기 위함
            driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/ul/li[1]/a')[0].click()
        except: # 첫 페이지 예외처리
            pass
        time.sleep(2)

        ########################## 1page 정보 수집과정 ##########################
        # 첫번째 정보 수집: 데이터 수집할 개수 정의 후 크롤링
        rows = len(driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[3]/table/tbody/tr'))
        df1_1 = first_info_crawling(rows)

        # 두번째 정보 수집: 기록보기 화살표 클릭 후 크롤링
        driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div[2]/a[2]')[0].click()
        df1_2 = second_info_crawling(rows)
        time.sleep(2)

        # 세번째 정보 수집: 세부기록 클릭 후 크롤링
        driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div[1]/ul/li[2]/a')[0].click()
        df1_3 = third_info_crawling(rows)
        time.sleep(2)

        # 1page 선수의 데이터 합치기
        df1 = pd.concat([df1_1, df1_2, df1_3], axis=1)
        df1


        ########################## 2page 정보 수집과정 ##########################
        # 2page 유무 확인 후 있으면 수집
        if len(driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[3]/div/a[4]/img')) != 0:

            # 첫번째 정보 수집: 기본기록 클릭 및 2page 클릭 및 데이터 수집할 개수 정의 후 크롤링
            driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/ul/li[1]/a')[0].click()
            time.sleep(2)
            driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[3]/div/a[4]/img')[0].click()
            time.sleep(2)
            rows = len(driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[3]/table/tbody/tr'))
            df2_1 = first_info_crawling(rows)

            # 두번째 정보 수집: 기록보기 화살표 클릭 및 2page 클릭 후 크롤링
            ### 기록보기 화살표 클릭
            driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div[2]/a[2]')[0].click()
            time.sleep(2)
            driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[3]/div/a[4]/img')[0].click()
            time.sleep(2)
            df2_2 = second_info_crawling(rows)

            # 세번째 정보 수집: 세부기록 클릭 후 2page 클릭 후 크롤링
            driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div[1]/ul/li[2]/a')[0].click()
            time.sleep(2)
            driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[3]/div/a[4]/img')[0].click()
            time.sleep(2)
            df2_3 = third_info_crawling(rows)

            # 2page 선수의 데이터 합치기
            df2 = pd.concat([df2_1, df2_2, df2_3], axis=1)

            # 1page, 2page 선수의 데이터 합치기
            df = pd.concat([df1, df2]).reset_index(drop=True)

        else:
            df = df1.copy()

        # 최종 저장
        if f'{BOX_TEAM_PATH}' not in os.listdir('선수데이터/타자/'):
            os.mkdir(f'선수데이터/타자/{BOX_TEAM_PATH}')

        df.to_csv(f'선수데이터/타자/{BOX_TEAM_PATH}/타자정보_{BOX_TEAM_PATH}({BOX_YEAR_PATH}).csv', index=False, encoding='utf-8-sig')
        print(f'=== {BOX_TEAM_PATH}({BOX_YEAR_PATH}) 저장완료 === 경과시간: {round(time.time()-t1)}초')
```
