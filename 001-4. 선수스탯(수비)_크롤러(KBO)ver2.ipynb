# Load Package

```python
import time
import os
import warnings

import pandas as pd 

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

warnings.filterwarnings(action='ignore')

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("user-agent=Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36")
```

# 크롤링

```python
t1 = time.time()

########################## 접속과정 ##########################
# page에 접속
driver = webdriver.Chrome('chromedriver_93.exe', chrome_options=options) # 크롬 드라이버 다운로드후 본인 경로로 줍니다.
base_url = 'https://www.koreabaseball.com/Record/Player/Defense/Basic.aspx'
driver.get(base_url)
time.sleep(2)


########################## 연도 변경과정 ##########################
# 년도 바꾸기 (option[2] ~ option[20]) 
for INDEX_YEAR in [num for num in range(20, 1, -1)]: 
    
    ### 년도 콤보박스 클릭
    driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[1]/select')[0].click()
    time.sleep(1)
    
    ### 세부년도 클릭
    BOX_YEAR = driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[1]/div[1]/select/option[{INDEX_YEAR}]')[0]
    BOX_YEAR_PATH = BOX_YEAR.text
    BOX_YEAR.click() # 2020년이 options[20]
    time.sleep(3)
    
    
    ### 제일 첫 페이지로
    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[1]/img')[0].click()
    time.sleep(3)
    

    ########################## 페이지 초기화 ##########################
    page = 0


    ########################## 정보수집 과정 ##########################
    while True:

        # 페이지 추가
        page += 1

        # 크롤링
        rows = len(driver.find_elements_by_xpath('/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr'))

        PLAYER = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[2]/a')[0].text for row in range(1, rows+1)]
        ID = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[2]/a')[0].get_attribute('href').split('=')[1] for row in range(1, rows+1)]
        TEAM = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[3]')[0].text for row in range(1, rows+1)]
        POS = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[4]')[0].text for row in range(1, rows+1)]
        G = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[5]')[0].text for row in range(1, rows+1)]
        GS = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[6]')[0].text for row in range(1, rows+1)]
        IP = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[7]')[0].text for row in range(1, rows+1)]
        E = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[8]')[0].text for row in range(1, rows+1)]
        PKO = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[9]')[0].text for row in range(1, rows+1)]
        PO = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[10]')[0].text for row in range(1, rows+1)]
        A = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[11]')[0].text for row in range(1, rows+1)]
        DP = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[12]')[0].text for row in range(1, rows+1)]
        FPCT = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[13]')[0].text for row in range(1, rows+1)]
        PB = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[14]')[0].text for row in range(1, rows+1)]
        SB = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[15]')[0].text for row in range(1, rows+1)]
        CS = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[16]')[0].text for row in range(1, rows+1)]
        CS_PERCENTAGE = [driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/table/tbody/tr[{row}]/td[17]')[0].text for row in range(1, rows+1)]

        # 데이터프레임 만들기
        df = pd.DataFrame({'선수명' : PLAYER,
                           'ID' : ID,
                           '팀명' : TEAM,
                           '포지션' : POS,
                          '경기' : G,
                          '선발경기' : GS,
                          '수비이닝' : IP,
                          '실책' : E,
                          '견제사' : PKO,
                          '자살' : PO,
                          '보살' : A,
                          '병살' : DP,
                          '수비율' : FPCT,
                          '포일' : PB,
                          '도루허용' : SB,
                          '도루실패' : CS,
                          '도루저지율' : CS_PERCENTAGE})

        # 데이터프레임 추가
        if page == 1:
            df_final = df.copy()
        elif page > 1:
            df_final = pd.concat([df_final, df]).reset_index(drop=True)


        ########################## 페이지 넘기기 과정 ##########################
        if len(driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a')) != 9:

            ### 1~5 구간
            if driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[2]')[0].text == '1':   
                if page % 5 == 1:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[3]')[0].click()
                elif page % 5 == 2:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[4]')[0].click()
                elif page % 5 == 3:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[5]')[0].click()
                elif page % 5 == 4:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[6]')[0].click()
                elif page % 5 == 0:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[7]')[0].click()

            ### 마지막 구간
            else:
                last_idx = len(driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a')) - 3
                if last_idx == 5:
                    last_idx = 0
                    
                # 최종 저장
                if page % 5 == last_idx:
                    df_final.to_csv(f'선수데이터/수비/수비정보_팀전체({BOX_YEAR_PATH}).csv', index = False, encoding = 'utf-8-sig')
                    print(f'=== 수비정보_팀전체({BOX_YEAR_PATH}) 저장완료 === 경과시간: {round(time.time()-t1)}초')
                    break


                if page % 5 == 1:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[4]')[0].click()
                elif page % 5 == 2:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[5]')[0].click()
                elif page % 5 == 3:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[6]')[0].click()
                elif page % 5 == 4:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[7]')[0].click()
                elif page % 5 == 0:
                    driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[8]')[0].click()


        ### 그 외 중간구간
        else:
            if page % 5 == 1:
                driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[4]')[0].click()
            elif page % 5 == 2:
                driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[5]')[0].click()
            elif page % 5 == 3:
                driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[6]')[0].click()
            elif page % 5 == 4:
                driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[7]')[0].click()
            elif page % 5 == 0:
                driver.find_elements_by_xpath(f'/html/body/form/div[3]/section/div/div/div[2]/div[3]/div[2]/div/a[8]')[0].click()

        time.sleep(3)
```
