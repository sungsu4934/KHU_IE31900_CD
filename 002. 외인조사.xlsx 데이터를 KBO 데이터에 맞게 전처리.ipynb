```python
def change_teamname1(x):  # 외인조사.xlsx
    
    """
    기아 -> KIA
    히어로즈, 넥센, 키움 -> 히어로즈/넥센/키움
    kt -> KT
    우리, 현대 -> 우리/현대
    """
    
    if x == '기아':
        x = 'KIA'
    elif x == 'kt':
        x = 'KT'
    elif (x == '히어로즈') or (x == '넥센') or (x == '키움') or (x == '우리'):
        x = '우리/히어로즈/넥센/키움'
        
    return x


def change_teamname2(x):  # KBO데이터
    
    """
    히어로즈, 넥센, 키움 -> 히어로즈/넥센/키움
    """
    
    if (x == '히어로즈') or (x == '넥센') or (x == '키움') or (x == '우리'):
        x = '우리/히어로즈/넥센/키움'
        
    return x
```

# 2.  수비데이터 전처리

```python
# 데이터 통합

for year in range(2020, 2001, -1):
    
    tmp = pd.read_csv(f'선수데이터/수비/수비정보_팀전체({year}).csv')
    tmp['연도'] = year
    tmp = tmp[['선수명', 'ID', '연도', '팀명', '포지션', '경기', '선발경기', '수비이닝', '실책', '견제사', '자살', '보살', '병살', '수비율', '포일', '도루허용', '도루실패', '도루저지율']]
    if year == 2020:
        defense = tmp.copy()
    else:
        defense = pd.concat([defense, tmp]).reset_index(drop=True)
        
defense
```

```python
YEARS = list(range(2020, 2001, -1))
TEAMS = ['한화', 'KIA', 'LG', 'SK', '롯데', '삼성', '두산', '우리/히어로즈/넥센/키움', 'NC', 'KT', '현대']

for YEAR in YEARS:
    print('')
    
    for TEAM in TEAMS:
        
        # 세부이름들 추출
        names = []
        foreigners = df.loc[(df['연도'] == YEAR) & (df['팀명'] == TEAM)]
        for foreigner in foreigners['선수명']:
            name = foreigner.split(' ')
            names.extend(name)

        # 삭제
        remove_index = defense.loc[(defense['연도'] == YEAR) & (defense['팀명'] == TEAM) & (defense['선수명'].isin(names))].index
        remove_players = list(defense.loc[remove_index, '선수명'].unique())
        defense.drop(remove_index, inplace=True)
        
        if len(foreigners["선수명"].unique()) != len(remove_players):
            print(f'{TEAM}({YEAR})의 {remove_players} ({len(remove_players)}명) 삭제 완료 == 삭제대상: {list(foreigners["선수명"].unique())}')
```
